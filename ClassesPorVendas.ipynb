### Modificando tipagens

```python
# removendo cifrão do valor da venda
df['Revenues'] = df['Revenues'].apply(lambda x: x.replace('$',''))
df
```

```python
df = df.astype({'Date' : 'datetime64[s]', 'Store ID' : 'str', 'Category' : 'str', 'Status' : 'str', 'Payment Type' : 'str', 'Place or Sale' : 'str', 'Product' : 'str', 'Product ID' : 'str', 'Payment Method': 'str', 'Revenues' : 'float', 'Products' : 'str'})
df.info()
```

### Removendo personalizações e tickets com receita zerada

```python
maskZeroRevenue = df['Revenues'] == 0
df = df[~maskZeroRevenue]
df
```
